```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, y_true = make_blobs(n_samples=500, centers=3,cluster_std = 0.06 ,random_state=42)
```

```python
df = pd.DataFrame(X , columns=['Feature1','Feature2'])
```

```python
scaler = StandardScaler()
df_scaler = scaler.fit_transform(df)
```

```python
wcss = []
k_range = range(1,11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters = k , random_state=42)
    kmeans.fit(df_scaler)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(k_range , wcss , marker='o')
```

```python
kmeans_final = KMeans(n_clusters = 6 , random_state=42)
```

```python
cluster_label = kmeans_final.fit_predict(df_scaler)
cluster_label
```

```python
df['cluster'] = kmeans_final
```

```python
sns.scatterplot(x=df['Feature1'] , y=df['Feature2'] , hue=df['cluster'], palette='viridis' )
```
