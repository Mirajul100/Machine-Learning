# Data cleaning and Data preprocessing

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df.dtypes
```

# Level Encoding 

```python
df_clean = df
```

```python
df_clean.head(10)
```

```python
df_clean.columns
```

```python
df_clean['sex'].value_counts()
```

```python
df_clean['sex'] = df_clean['sex'].map({'male':0 , 'female':1})
```

```python
df_clean
```

```python
df_clean['smoker'].value_counts()
```

```python
df_clean['smoker'] = df_clean['smoker'].map({'no':0 ,'yes':1})
```

```python
df_clean
```

```python
df_clean.rename(columns = {
    'sex':'is_female',
    'smoker':'is_smoker'
} , inplace = True)
```

```python
df_clean.head(10)
```

# One hot encoding

```python
df_clean.columns
```

```python
df_clean['region'].value_counts()
```

```python
df_clean = pd.get_dummies(df_clean , columns = ['region'] , drop_first = True)
```

```python
df_clean
```

```python
df_clean = df_clean.astype(int)
```

```python
df_clean.head(10)
```

```python
df_encoded = df_clean
```

```python
df_encoded
```

# Feature engineering and Extraction

```python
df_encoded['bmi_category'] = pd.cut(
    df_encoded['bmi'],
    bins = [0, 18.5, 24.9, 29.9, float('inf')],
    labels = ['UnderWeight', 'Normal', 'OverWeight', 'Obese']
)
```

```python
df_encoded
```

```python
df_encoded['bmi_category'].value_counts()
```

```python
df_encoded = pd.get_dummies(df_encoded, columns=['bmi_category'], drop_first = True)
```

```python
df_encoded = df_encoded.astype(int)
```
