```python
def ploting (var , num):
    plt.subplot(3 ,3 , num)
    sns.histplot(df[var] , kde = True)

ploting('age' , 1)
ploting('daily_social_media_hours' , 2)
ploting('sleep_hours' , 3)
ploting('screen_time_before_sleep' , 4)
ploting('academic_performance' , 5)
ploting('physical_activity' , 6)


plt.tight_layout()
```

```python
def boxplot(var):
    plt.figure(figsize=(6,4))
    sns.boxenplot(df[var])
boxplot('age' , 1)
boxplot('daily_social_media_hours')
boxplot('platform_usage')
boxenplot('sleep_hours')
boxenplot('screen_time_before_sleep')
boxenplot('physical_activity')
boxenplot('stress_level')
boxenplot('anxiety_level')
```
